```python
# Importing Libraries
import ast
import pandas as pd
import seaborn as sns
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

Filter for INDIA D.A Roles

```python
df_DA_IND= df[(df['job_country'] == 'India') & (df['job_title_short'] == 'Data Analyst')]
```

```python
df_DA_IND
```

```python
df_plot = df_DA_IND['job_location'].value_counts().head(10).to_frame()
sns.set_theme(style='ticks')
sns.barplot(data=df_plot, x='count', y='job_location', hue='count', palette='dark:b_r', legend=False)
sns.despine()
plt.title('Counts of Job Location for Data Analyst in India')
plt.xlabel('Number of Jobs')
plt.ylabel('')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Your dictionary of columns and titles
dict_column = {
    'job_work_from_home': 'Work from Home Offered',
    'job_no_degree_mention': 'Degree Requirement',
    'job_health_insurance': 'Health Insurance Offered'
}

fig, ax = plt.subplots(1, 3, figsize=(12, 5))

for i, (column, title) in enumerate(dict_column.items()):
    # Get the value counts
    value_counts = df_DA_IND[column].value_counts()
    # Use the unique values as labels
    labels = value_counts.index.tolist()
    ax[i].pie(value_counts, labels=labels, autopct='%1.1f%%', startangle=90)
    ax[i].set_title(title)

# Uncomment if you want to add a suptitle
# plt.suptitle('Benefit Analysis of Data Jobs', fontsize=16)
plt.show()
```

```python
df_plot = df_DA_IND['company_name'].value_counts().head(10).to_frame()

sns.set_theme(style='ticks')
sns.barplot(data=df_plot, x='count', y='company_name', hue='count', palette='dark:b_r', legend=False)
sns.despine()
plt.title('Counts of Companies for Data Analyst in the US')
plt.xlabel('Number of Jobs')
plt.ylabel('')
plt.show()
```
